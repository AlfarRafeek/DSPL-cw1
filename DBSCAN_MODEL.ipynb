```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
Data_path = "/content/preprocessed_train_data.csv"
df = pd.read_csv(Data_path, low_memory=False)

# Select numerical features for clustering
X = df.select_dtypes(include="number").dropna()

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
df.loc[X.index, 'cluster'] = dbscan.fit_predict(X)
```

```python
print("Missing Values:\n", X.isnull().sum())
print("Data Types:\n", X.dtypes)
print("Shape of Data:", X.shape)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['luxury_sales'], y=df['fresh_sales'], hue=df['cluster'], palette='viridis', legend='full')
plt.title("DBSCAN Clustering Results")
plt.show()
```

```python
unique_clusters = df['cluster'].nunique()
print(f"Number of clusters found: {unique_clusters}")
print(df['cluster'].value_counts())
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
Data_path = "/content/preprocessed_train_data.csv"
df = pd.read_csv(Data_path, low_memory=False)

# Select numerical features for clustering
X = df.select_dtypes(include="number").dropna()

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
df.loc[X.index, 'cluster'] = dbscan.fit_predict(X)

print("Missing Values:\n", X.isnull().sum())
print("Data Types:\n", X.dtypes)
print("Shape of Data:", X.shape)

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['luxury_sales'], y=df['fresh_sales'], hue=df['cluster'], palette='viridis', legend='full')
plt.title("DBSCAN Clustering Results")
plt.show()

unique_clusters = df['cluster'].nunique()
print(f"Number of clusters found: {unique_clusters}")
print(df['cluster'].value_counts())
```

```python
plt.figure(figsize=(15, 10))

# Scatter plot for luxury_sales vs fresh_sales
plt.subplot(2, 2, 1)
sns.scatterplot(x=df['luxury_sales'], y=df['fresh_sales'], hue=df['cluster'], palette='viridis', legend='full')
plt.title("DBSCAN: Luxury Sales vs Fresh Sales")

# Scatter plot for luxury_sales vs dry_sales
plt.subplot(2, 2, 2)
sns.scatterplot(x=df['luxury_sales'], y=df['dry_sales'], hue=df['cluster'], palette='viridis', legend='full')
plt.title("DBSCAN: Luxury Sales vs Dry Sales")

# Scatter plot for fresh_sales vs dry_sales
plt.subplot(2, 2, 3)
sns.scatterplot(x=df['fresh_sales'], y=df['dry_sales'], hue=df['cluster'], palette='viridis', legend='full')
plt.title("DBSCAN: Fresh Sales vs Dry Sales")

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot in 3D space
ax.scatter(df['luxury_sales'], df['fresh_sales'], df['dry_sales'], c=df['cluster'], cmap='viridis', marker='o')

# Labels
ax.set_xlabel('Luxury Sales')
ax.set_ylabel('Fresh Sales')
ax.set_zlabel('Dry Sales')
ax.set_title("DBSCAN Clustering (3D View)")

plt.show()
```
